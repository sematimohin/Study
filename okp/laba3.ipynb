```python
import numpy as np
import matplotlib.pyplot as plt

Esi_minus10 = [4,
5.25,
6.25,
7.1,
8.1,
9.25,
10.5,
11.5,
12.1,
]

Esi_plus10 = [5.5,
7,
7.5,
9.5,
10.75,
12.25,
13.75,
14.75,
15.5,
]

Ek_minus10 = [15,
14.75,
12.25,
10.25,
9,
8,
7,
6.5,
6.1,
]

Ek_plus10 = [13,
10.75,
9,
7.6,
6.8,
6,
5.5,
5,
4.9,
]

R6 = np.arange(0,4.5,0.5)

plt.figure(figsize=(8, 4), dpi=120)
plt.plot(R6, Esi_minus10, label='Eси(-10%)', color='blue', linestyle='--')
plt.plot(R6, Esi_plus10, label='Eси(+10%)', color='green', linestyle=':')
plt.plot(R6, Ek_minus10, label='Eк(-10%)', color='red', linestyle='-')
plt.plot(R6, Ek_plus10, label='Eк(+10%)', color='orange', linestyle='-.')

plt.scatter(R6, Esi_minus10, color='blue', s = 20)
plt.scatter(R6, Esi_plus10, color='green', s = 20)
plt.scatter(R6, Ek_minus10, color='red', s = 20)
plt.scatter(R6, Ek_plus10, color='orange', s = 20)


plt.xlabel('R5, R6, кОм')
plt.ylabel('Eк,Еси, В')
plt.legend(bbox_to_anchor=(0, -0.15), loc='upper left', prop={'size': 10})
plt.grid(True)
plt.show()
```
