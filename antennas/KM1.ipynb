# Задание 1
Вывеcти формулу для диаграммы направленности __горизонтального электрического__ диполя. Диполь расположен на заданной высоте __h/λ = 0.25__ над металлической плоскостью. Определить плоскости Е и Н. Рассчитать и построить диаграмму направленности. 

Рассмотрим горизонтальный электрический диполь, ориентированный вдоль оси X, с током I, и расположенный на высоте h над идеально проводящей плоскостью.  `Как известно, на поверхности проводящей плоскости
выполнено граничное условие $E(\tau) = 0$.`

Это условие автоматически выполнено в системе, 
состоящей из двух диполей, заданного и его зеркального изображения в плоскости
металлического листа, причём, `направление тока в зеркально отражённом диполе противоположно току в заданном диполе. Таким образом, влияние плоскости заменяется на влияние "отражённого" диполя`.

Далее будет объяснение из chatGPT, т.к. из лекций я ничего не понял)

Шаги решения:
1. __Модель горизонтального диполя над плоскостью__: Когда диполь находится над металлической плоскостью, появляется его зеркальное отражение, что эквивалентно расположению двух диполей: один на высоте h, а другой — на той же высоте ниже плоскости, образуя зеркальный диполь с противоположной ориентацией.

2. __Формула для диаграммы направленности__: Диаграмма направленности горизонтального диполя на высоте 
ℎ=0.25𝜆 над металлической плоскостью может быть определена на основе поля диполя и его зеркального изображения. Рассматривая интерференцию между этими источниками, интенсивность поля в зависимости от угла θ определяется как:




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Определим параметры
h_lambda = 0.25  # высота диполя над плоскостью в долях длины волны
d = 2 * h_lambda  # расстояние между диполем и его изображением
k = 2 * np.pi  # волновое число (для длины волны lambda)

# Определим углы theta от 0 до 180 градусов (в радианах)
theta = np.linspace(0, np.pi*2, 360)

# Вычислим диаграмму направленности
E_theta_horizontal = np.abs(np.sin(theta) * (1 + np.exp(1j * k * d * np.cos(theta))))
E_theta_horizontal_norm = E_theta_horizontal/np.max(E_theta_horizontal) # нормируем массив

# Построим диаграмму направленности
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)
ax.plot(theta, E_theta_horizontal_norm, label=r'$E(\theta)$')

# Настройки графика
plt.title("Нормированная диаграмма направленности горизонтального диполя")
plt.legend(loc="upper right")
plt.show()

```

# Задание 2

Вывеcти формулу для диаграммы направленности __вертикального электрического__ диполя. Диполь расположен на заданной высоте __h/λ = 0.75__ над металлической плоскостью. Определить плоскости Е и Н. Рассчитать и построить диаграмму направленности. 

```python
# Параметры для вертикального диполя
h_lambda_vertical = 0.75  # высота диполя над плоскостью в долях длины волны
d_vertical = 2 * h_lambda_vertical  # расстояние между диполем и его изображением

# Вычислим диаграмму направленности для вертикального электрического диполя
E_theta_vertical = np.abs(np.cos(theta) * (1 + np.exp(1j * k * d_vertical * np.cos(theta))))
E_theta_vertical_norm = E_theta_vertical/np.max(E_theta_vertical) # нормировка

# Построим диаграмму направленности для вертикального диполя
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)
ax.plot(theta, E_theta_vertical_norm, label=r'$E(\theta)$ (вертикальный диполь)', color='g')

# Настройки графика
plt.title("Нормированная диаграмма направленности вертикального электрического диполя")
plt.legend(loc="upper right")
plt.show()
```

# Задание 3

Всё тоже самое, но диполь магнитный вертикальный на высоте 0.25

```python
# Пересчитываем диаграмму направленности для вертикального магнитного диполя
H_theta_corrected = np.abs(np.sin(theta) * (1 - np.exp(1j * k * d * np.cos(theta))))

# Построим скорректированную диаграмму направленности для вертикального магнитного диполя
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)
ax.plot(theta, H_theta_corrected, label=r'$H(\theta)$ (вертикальный магнитный диполь)', color='r')

# Настройки графика
plt.title("Диаграмма направленности вертикального магнитного диполя (исправленная)")
plt.legend(loc="upper right")
plt.show()
```

```python
# Пересчитываем диаграмму направленности с учетом поправок
H_theta_final = np.abs(np.sin(theta) * (1 - np.exp(1j * k * d * np.cos(theta))))

# Построим финальную диаграмму направленности для вертикального магнитного диполя
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)
ax.plot(theta, H_theta_final, label=r'$H(\theta)$ (вертикальный магнитный диполь)', color='r')

# Настройки графика
plt.title("Финальная диаграмма направленности вертикального магнитного диполя")
plt.legend(loc="upper right")
plt.show()
```
