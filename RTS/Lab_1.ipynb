```python
import matplotlib.pyplot as plt

#1 положение: Двигаю первый маяк по координате y1 от 0.5 до 3.5 c шагом 0.5

# Данные
delta = [0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0]
m_x = [-0.017, -1.989, 0.249, 6.375, 13.67, 22.837, -76.77]
m_y = [0.240, 47.868, 103.090, 141.89, 170.44, 193.020, 88.177]
sigma_x = [3.938, 3.676, 3.689, 3.998, 4.321, 4.679, 6.728]
sigma_y = [4.476, 5.928, 5.220, 4.241, 3.860, 3.806, 5.100]

# Создание фигуры с двумя субплотами
plt.figure(figsize=(12, 8))

# График для математических ожиданий
plt.subplot(2, 1, 1)
plt.plot(delta, m_x, marker='*', linestyle='none',ms = 10, color='blue', label='$M_x$')
plt.plot(delta, m_y, marker='o', linestyle='none',ms = 8, color='red', label='$M_y$')
plt.xlabel(r'$\Delta$, м', fontsize=12)
plt.ylabel('МО, см', fontsize=12)
plt.legend()
plt.grid(True)

# График для СКО
plt.subplot(2, 1, 2)
plt.plot(delta, sigma_x, marker='*', linestyle='none',ms = 10, color='blue', label=r'$\sigma_x$')
plt.plot(delta, sigma_y, marker='o', linestyle='none',ms = 8, color='red',label=r'$\sigma_y$')
plt.xlabel(r'$\Delta$, м', fontsize=12)
plt.ylabel('СКО, см', fontsize=12)
plt.legend()
plt.grid(True)

# Настройка отступов и отображение
plt.tight_layout()
plt.show()
```
