```python
import numpy as np
import matplotlib.pyplot as plt

# Углы для построения диаграмм
theta = np.linspace(0, np.pi, 360)
phi = np.linspace(0, 2*np.pi, 360)

# Диаграмма направленности D(θ) в плоскости X0Z и Y0Z
D_theta = np.sin(theta)

# Диаграмма направленности в плоскости XOY (зависит от phi)
D_phi = np.sin(phi)

# Функция для построения диаграммы направленности
def plot_polar(theta, D, title):
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    ax.plot(theta, D)
    ax.set_title(title)
    plt.show()

# Диаграмма направленности в плоскости X0Z (φ = 0)
plot_polar(theta, D_theta, "Диаграмма направленности в плоскости X0Z")

# Диаграмма направленности в плоскости Y0Z (φ = pi/2)
plot_polar(theta, D_theta, "Диаграмма направленности в плоскости Y0Z")

# Диаграмма направленности в плоскости XOY (θ = pi/2)
plot_polar(phi, D_phi, "Диаграмма направленности в плоскости XOY")
```
