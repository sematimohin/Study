```python
import numpy as np
import matplotlib.pyplot as plt

#    
theta = np.linspace(0, np.pi, 360)
phi = np.linspace(0, 2*np.pi, 360)

#   D()   X0Z  Y0Z
D_theta = np.sin(theta)

#     XOY (  phi)
D_phi = np.sin(phi)

#     
def plot_polar(theta, D, title):
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    ax.plot(theta, D)
    ax.set_title(title)
    plt.show()

#     X0Z ( = 0)
plot_polar(theta, D_theta, "    X0Z")

#     Y0Z ( = pi/2)
plot_polar(theta, D_theta, "    Y0Z")

#     XOY ( = pi/2)
plot_polar(phi, D_phi, "    XOY")
```
